```python
#api
# É um conjunto de rotinas e padrões estabelecidos por um software para a utilização
# das suas funcionalidades por aplicativos, não importando sua implementação, apenas 
# seu serviço. 

# biblioteca request
# é uma requisição http. 

import requests 
import json
import pandas as pd   
import seaborn as sns 
import matplotlib.pyplot as plt 
from wordcloud import WordCloud 
from pprint import pprint

# a proposta dos dados legistativos abertos é a transparência das informações da 
# câmara de deputados para todos. 

# requisição dos dados dos deputados 
url = 'https://dadosabertos.camara.leg.br/api/v2/deputados'
# parametros = {'id': 204549}
parametros = {}
resposta = requests.request("GET", url, params= parametros)
objeto = json.loads(resposta.text)
dados_parametros = objeto['dados']

pprint(dados_parametros)

dp = pd.DataFrame(dados_parametros)

dp.info()

# analisando os dados da api 
# quantidade de deputados por uf 

x = dp['siglaUf'].value_counts().head(10).index
y = dp['siglaUf'].value_counts().head(10).values

plt.figure(figsize= (10,5))

sns.barplot(x=x, y=y, palette="deep")
plt.plot()
```

```python
# quantidade de deputados por uf 

x = dp['siglaPartido'].value_counts().head(10).values
y = dp['siglaPartido'].value_counts().head(10).index

plt.figure(figsize= (10,5))

sns.barplot(x=x, y=y, palette="deep")
plt.plot()
```
